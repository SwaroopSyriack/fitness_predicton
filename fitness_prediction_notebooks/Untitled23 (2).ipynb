```python
def capped_cumsum(series, cap=20):
    cumsum = 0
    result = []
    for value in series:
         if cumsum + value <= cap:
             cumsum += value
             result.append(value)
    return sum(result)
```

```python
aggregation_functions = {
    'steps_per_calorie': 'sum',  # Sum numeric columns
    'distance_km': 'sum',
    'active_minutes': lambda x: capped_cumsum(x),  # Use the capped cumulative sum function
    'sleep_hours': 'mean',       # Average sleep
    'heart_rate_avg': 'mean',    # Average heart rate
    'workout_type': lambda x: x.iloc[0],  # Take the first occurrence
    'weather_conditions': lambda x: x.mode()[0],  # Most frequent weather condition
    'location': lambda x: x.mode()[0],     # Most frequent location
    'mood': lambda x: x.mode()[0]          # Most frequent mood
}
```

```python
df_grouped = df_new.groupby(['user_id','date']).agg(aggregation_functions).reset_index()
```

```python
a = df_new[(df_new['mood'] == "Happy")]
b = df_new[(df_new['mood'] == "Tired")]
c = df_new[(df_new['mood'] == "Neutral")]
d = df_new[(df_new['mood'] == "Stressed")]
```

# **resampling**

```python
from sklearn.utils import resample

data_A_resampled = resample(a , replace=True, n_samples=75000, random_state=0)
data_B_resampled = resample(b, replace=True, n_samples=75000, random_state=0)
data_C_resampled = resample(c , replace=True, n_samples=75000, random_state=0)
data_D_resampled = resample(d, replace=True, n_samples=75000, random_state=0)

# Concatenate resampled data
resampled_data = pd.concat([data_A_resampled,data_B_resampled,data_C_resampled,data_D_resampled])
```

```python
x = resampled_data.iloc[:,:-1]
```

```python
y = resampled_data.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
cols = ['day_of_week','workout_type', 'weather_conditions', 'location']
for col in cols:
    x[col] = label_encoder.fit_transform(x[col])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

columsn_scale = ['steps_per_calorie','distance_km','active_minutes','sleep_hours','heart_rate_avg']
# Step 4: Fit and transform the data
x[columsn_scale] = scaler.fit_transform(resampled_data[columsn_scale])
```

```python
# Get unique users
unique_users = data['user_id'].unique()

# Dictionary to store accuracy for each user and model
user_model_accuracy = {}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Loop over each unique user and train models
for user in unique_users:
    # Filter data for the specific user
    user_data = x[x['user_id'] == user]

    # Check if there is enough data to train the model
    if len(user_data) < 2:
        continue  # Skip this user if there's insufficient data



    # Split the data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

    # Define the models to test
    models = {
        'DecisionTree': DecisionTreeClassifier(),
        'RandomForest': RandomForestClassifier(n_estimators=10),
        'SVC': SVC()
    }

    # Initialize a dictionary to hold accuracy scores for the current user
    accuracy_scores = {}
    # Train and evaluate each model
    for model_name, model in models.items():
        model.fit(x_train, y_train)  # Fit model on training data
        y_pred = model.predict(x_test)  # Make predictions on test data
        accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy

        # Store the accuracy score for the current model
        accuracy_scores[model_name] = accuracy

    # Store the accuracy scores for the current user
    user_model_accuracy[user] = accuracy_scores

# Convert to DataFrame for easier viewing
accuracy_df = pd.DataFrame(user_model_accuracy).T.fillna(0)  # Transpose and fill NaN with 0
accuracy_df.index.name = 'user_id'  # Name the index for clarity

# Print the accuracy DataFrame
print(accuracy_df)
```
