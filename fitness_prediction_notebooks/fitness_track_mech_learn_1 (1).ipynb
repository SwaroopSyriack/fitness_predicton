```python
from sklearn.preprocessing import LabelEncoder

label_encoder=LabelEncoder()

df[['date']]=df[['date']].apply(LabelEncoder().fit_transform)
```

```python
df
```

```python
df1=pd.get_dummies(df[['workout_type','location','weather_conditions','mood']],drop_first=True)
```

```python
df1.head()
```

```python
df2=pd.concat([df,df1],axis=1)
```

```python
df2.head()
```

```python
df2.columns
```

```python
df3=df2.drop(['workout_type','location','weather_conditions','mood'],axis=1)
```

```python
df3.columns
```

```python
x = df3.iloc[:,0:-3].values
x
```

```python
y = df3.iloc[:,-3].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
print("x_train: ",x_train.shape)
print("y_train: ",y_train.shape)
print("x_test: ",x_test.shape)
print("y_test: ",y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators = 30, criterion = 'entropy', random_state = 42)
```

```python
model_rf.fit(x_train,y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
```

```python
y_pred_rf
```

```python
print("Accuracy : ", accuracy_score(y_test, y_pred_rf))
```
